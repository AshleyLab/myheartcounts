```r
rm(list=ls())
library(ggplot2)
```

```r
hk_data=read.table("../../../data/tables/cardiovascular-motionActivityCollector-v1.tsv",header=TRUE,sep='\t')
```

```r
tail(hk_data)
```

```r
hk_data$uploadDate=as.Date(hk_data$uploadDate)
```

```r
min(hk_data$uploadDate)
```

```r
max(hk_data$uploadDate)
```

```r
library(repr)
options(repr.plot.width=10, repr.plot.height=4)
ggplot(data=hk_data,
      aes(x=hk_data$uploadDate))+
geom_histogram(bins=100)+
xlab("uploadDate")+
ylab("motionActivityCollector")+
theme_bw(20)+
ggtitle(paste(min(hk_data$uploadDate),'-',max(hk_data$uploadDate)))
```

```r
hk_data$createdOn=as.Date(hk_data$createdOn)
```

```r
ggplot(data=hk_data,
      aes(x=createdOn,
         y=uploadDate))+
geom_point(alpha=0.01)+
xlab("createdOn")+
ylab("uploadDate")
```
